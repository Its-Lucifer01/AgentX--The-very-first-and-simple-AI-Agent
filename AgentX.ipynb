# Setting up memory

```python
import json
import os
```

```python
MEMORY_FILE = "memory.json"

# initialize memory file
if not os.path.exists(MEMORY_FILE):
    with open(MEMORY_FILE, "w") as f:
        json.dump([], f)

def store_memory(key, value):
    with open(MEMORY_FILE, "r") as f:
        data = json.load(f)

    data.append({"key": key, "value": value})

    with open(MEMORY_FILE, "w") as f:
        json.dump(data, f, indent=2)

def search_memory(key):
    with open(MEMORY_FILE, "r") as f:
        data = json.load(f)

    results = [item["value"] for item in data if key.lower() in item["key"].lower()]
    return results
```

# Adding Tools

```python
import random
```

```python
# calculator tool
def calculate(expression):
    try:
        return eval(expression)
    except:
        return "Error evaluating expression."

# weather tool
def weather_api(city):
    temp = random.randint(18, 35)
    return f"The temperature in {city} is {temp}°C."

# search tool
def search_api(query):
    return f"Mock search result for '{query}'. (In real use, call Google/Bing API)"
```

# Agent Brain

```python
def agent(query):

    query_lower = query.lower()

    #  Check for memory add
    if query_lower.startswith("remember"):
        # pattern: remember that my fav food is pizza
        parts = query.split("that")
        if len(parts) > 1:
            fact = parts[1].strip()
            store_memory(key=fact.split()[0], value=fact)
            return f"Got it! I'll remember that '{fact}'."
        else:
            return "What should I remember?"

    # Check memory recall
    if query_lower.startswith("what do you remember") or "remember" in query_lower:
        key = query_lower.replace("remember", "").strip()
        mem = search_memory(key)
        return mem if mem else "I don't have anything stored for that."

    # Calculator
    if any(c.isdigit() for c in query):
        try:
            result = calculate(query)
            return f"Calculation result: {result}"
        except:
            pass

    # Weather
    if "weather" in query_lower:
        city = query.split("in")[-1].strip()
        return weather_api(city)

    # Search
    if "search" in query_lower:
        q = query_lower.replace("search", "").strip()
        return search_api(q)

    # Default chat response
    return f"You said: {query}. I'm a simple agent but I'm learning!"
```
